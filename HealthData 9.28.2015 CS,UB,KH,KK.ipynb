```python
#Graph with summary statistics - I'm assuming this is just for Average Covered Charges
#What is the most expensive state for healthcare and what is the least expensive?
#Geographic plot by state of healthcare billing using Average Covered Costs.
#Sum and group by state
#Sum and group by DRG
#Use bokeh instead of matplotlib

#Prepare data for Summary Statistics - from last exercise
__author__ = 'Chris Sneed'
# Import section
import os
import numpy as np
import pandas as pd
import matplotlib as mpl
from bokeh.sampledata import us_states
from bokeh.plotting import figure, show, output_file

# Read in the health data
govdata = pd.read_csv ('C:\Users\W540\Jupyter\datagov.csv')

# Assign each column to a new data column without the $
acc = govdata[' Average Covered Charges '].str.lstrip('$')
atp = govdata[' Average Total Payments '].str.lstrip('$')
amp = govdata['Average Medicare Payments'].str.lstrip('$')

# Change the data type from oject to float so that it is numeric
# Assign to another new variable so the new data type "sticks"
accnum = acc.astype(float)
atpnum = atp.astype(float)
ampnum = amp.astype(float)

#Mean, Mode, Std. Dev for last 3 columns
accmean = accnum.mean()
accmode = accnum.mode()
accstd = accnum.std()

atpmean = atpnum.mean()
atpmode = atpnum.mode()
atpstd = atpnum.std()

ampmean = ampnum.mean()
ampmode = ampnum.mode()
ampstd = ampnum.std()

#update govdata table
govdata.loc[:,'Avg Covered Charges - Float'] = accnum
govdata.loc[:,'Avg Total Payments - Float'] = atpnum
govdata.loc[:,'Avg Medicare Payments - Float'] = ampnum

#Display results
print "Mean of Average Covered Charges" " = " "%.2f" % accmean
print "Mode of Average Covered Charges" " = " "%.2f" % accmode
print "STD of Average Covered Charges" " = " "%.2f" % accstd

print "Mean of Average Total Payments" " = " "%.2f" % atpmean
print "Mode of Average Total Payments" " = " "%.2f" % atpmode
print "STD of Average Total Payments" " = " "%.2f" % atpstd

print "Mean of Average Medical Payments" " = " "%.2f" % ampmean
print "Mode of Average Medical Payments"
print ampmode
print "STD of Average Medical Payments" " = " "%.2f" % ampstd
```

```python
#save new columns back to original csv
govdata.to_csv('C:\Users\W540\Jupyter\datagov.csv')
```

```python
#Convert Average Covered Charges ('Avg Covered Charges - Float') from float to integer
govdata['Avg Covered Charges - Int'] = govdata['Avg Covered Charges - Float'].astype('int')
```

```python
#Append new column to end of data set and write it back to original csv
govdata.loc[:,'Avg Covered Charges - Int'] = govdata['Avg Covered Charges - Int']
govdata.to_csv('C:\Users\W540\Jupyter\datagov.csv')
```

```python
#Group by state and sum the Average Covered Charges by state
#Sort from most expensive to least expensive
group_state = govdata.groupby(['Provider State',])
group_state['Avg Covered Charges - Int'].agg({'State Sum': np.sum, 'Mean': np.mean, 'Std Dev.': np.std}).sort('State Sum',ascending=[0])
```
